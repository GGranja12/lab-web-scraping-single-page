```python
#Importing relevant libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# get the link
link = "https://www.billboard.com/charts/hot-100/"
```

```python
response = requests.get(link)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
#print(soup.prettify())
```

```python
#getting the ranking position
ranking=[]
for position in soup.find_all(class_= "c-label a-font-primary-bold-l u-font-size-32@tablet u-letter-spacing-0080@tablet"):
    ranking.append(position.get_text(strip=True))    
```

```python
ranking
```

```python
titles=[]
for li in soup.select("li h3.c-title"):
    titles.append(li.get_text(strip=True))
titles
```

```python
#trying to get the artists list
artists=[]
for li in soup.select("li span.c-label"):
    artists.append(li.get_text(strip=True))
artists
```

```python
#defining our dict to get the dataframe
dict_data = {"ranking": ranking "titles": titles, "artists": artists}
```

```python
#getting the dataframe
df = pd.DataFrame(dict_data)
df
```
